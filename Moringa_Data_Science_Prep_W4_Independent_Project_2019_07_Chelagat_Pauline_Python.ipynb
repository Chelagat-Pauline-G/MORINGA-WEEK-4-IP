```
#loading our dataset and previewing the first 5 columns
df=pd.read_csv('autolib.csv')
df.head(5)
```

```
df.isnull().count()
```

```
#UNIFORMITY
#Changing the column names to lower case
#So that we can easily access the columns
df.columns=df.columns.str.lower()
df.head(3)
```

```
#VALIDITY
#dropping columns we dont need
df1=df.drop(['cars','charge slots', 'charging status','displayed comment','id','kind','geo point','scheduled at','station type', 'subscription status', 'year', 'month'], axis=1)
df1
```

```
#Selecting Paris
df2=df1.loc[df['city']=='Paris']
df2
```

```
#selecting only the stations where the rental status is operational
df3=df2.loc[df['rental status']=='operational']
df3
```

```
#selecting only the statiosn where the rental status is operational
df4=df3.loc[df['status']=='ok']
df4
```

```
#exporting the cleaned dataset
df4.to_csv('carhire.csv')
df4.head(15)
```

```
#checking whether the changes have been saved
df4.info()
```

```
#most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
#selecting where there was a bluecar counter
df4a=df4.loc[df['bluecar counter']>0]
df4a
#finding the most popular hour for the bluecar counter
df4b=df4a['hour'].value_counts()
df4b
```

```
#What is the most popular hour for returning cars?

df5=df4['hour'].value_counts()
df5
```

```
#What station is the most popular?
#overall

df6=df4['public name'].value_counts()
df6
```

```
#What station is the most popular?
#At the most popular picking hour?
df6a=df4.loc[df['hour']==21]
df6a

df6b=df6a.groupby(['postal code'])['public name'].value_counts().sort_values(ascending=False)
df6b
```

```
#What postal code is the most popular for picking up Blue cars? 
df8=df4a.groupby(['bluecar counter'])['postal code'].value_counts()
df8
```

```
#Does the most popular station belong to that postal code?
df8a=df4a.groupby(['postal code'])['public name'].value_counts().sort_values(ascending=False)
df8a
```

```
#Do the results change if you consider Utilib and Utilib 1.4 instead of Blue cars? 

df9=df4.loc[(df['utilib counter']>0) & (df['utilib 1.4 counter']>0)]
df9

      
```

```
#most popular hour of the day for picking up Utilib and Utilib 1.4 in the city of Paris over the month of April 2018.

#finding the most popular hour for Utilib and Utilib 1.4
df9a=df9.groupby(['utilib 1.4 counter', 'utilib counter'])['hour'].value_counts()
df9a
```

```
#What postal code is the most popular for picking up Utilib and Utilib 1.4? 
df9b=df9.groupby(['utilib 1.4 counter', 'utilib counter'])['postal code'].value_counts()
df9b
```

```
#Does the most popular station belong to that postal code?
df9c=df9.groupby(['postal code'])['public name'].value_counts().sort_values(ascending=False)
df9c


#combined.groupby(['date', 'product'])['product'].count().sort_values(ascending = False)
```
